```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
df = pd.read_csv('helix_doc.csv')
#df.head()
#df.dtypes
#len(df)
#np.mean(df.ra), np.mean(df.dec)
```

```python
ratio = df.f90/df.f140
df["ratio"] = ratio # add new column for ratio 
#df
```

```python
Temp = -57/(np.log(df.ratio*0.64**5))
df["Temp"] = Temp #add new column for Temp
```

```python
c = 3.0E+8 
h = 6.62E-34
kb = 1.38E-23
lam = 1.4E-4
a = np.exp(h*c/(lam*kb*df['Temp']))
b = 1/(a-1)
planck = h*c*b/lam**3
```

```python
df["planck"] = planck 
```

```python
su = df.f140*5.288*10**-29
D = 200*3.0857*10**16  #here distance is 200 pc
Mass = 0.4*su*D**2/df.planck
```

```python
df["Mass"]=Mass
```

```python
DF = pd.DataFrame(df)
DF.to_csv('helix_final.csv')
```

```python
from sklearn import preprocessing
```

```python
A = df.f100.values.reshape(-1,1)
I = df.f60.values.reshape(-1,1)
N = df.Temp.values.reshape(-1,1)
P = df.planck.values.reshape(-1,1)
B = df.Mass.values.reshape(-1,1)
```

```python
scaled_f100 = preprocessing.MinMaxScaler().fit(A).transform(A)
scaled_f60 = preprocessing.MinMaxScaler().fit(N).transform(N)
scaled_Temp = preprocessing.MinMaxScaler().fit(I).transform(I)
scaled_Planck = preprocessing.MinMaxScaler().fit(P).transform(P)
scaled_Mass = preprocessing.MinMaxScaler().fit(B).transform(B)
```

```python
df['scaled_f100'] = scaled_f100
df['scaled_f60'] = scaled_f60
df['scaled_Temp'] = scaled_Temp
df['scaled_Planck'] = scaled_Planck
df['scaled_Mass'] = scaled_Mass
df.head()
```
