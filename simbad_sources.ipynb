```python
df_star =df[df.OTYPE=='Star']
df_rgb =df[df.OTYPE=='RGB*']
df_gal= df[df.OTYPE=="Galaxy"]
df_hb=df[df.OTYPE=="HB*"]
df_radio =df[df.OTYPE=='Radio*']
df_pm=df[df.OTYPE=="PM*"]
df_wd_cand=df[df.OTYPE=="Candidate_WD*"]
df_fir=df[df.OTYPE=="FIR"]
df_radio = df[df.OTYPE=='Radio']
df_x =df[df.OTYPE=='X']
df_ttau = df[df.OTYPE=='TTau*']
df_orian = df[df.OTYPE=='Orian_V*']
df_radiosub = df[df.OTYPE=='Radio(sub-mm)']
df_cloud = df[df.OTYPE=='PartofCloud']

df_lmass = df[df.OTYPE=='low-mass*']
df_IR = df[df.OTYPE=='IR']
df_dense = df[df.OTYPE=='denseCore']
df_HH = df[df.OTYPE=='HH']
```

```python
plt.figure(figsize=[10,8])
plt.plot(df_star['ra'], df_star['dec'],'*b', label= 'Star')
plt.plot(df_rgb['ra'], df_rgb['dec'],'Xr', label= 'RGB')
plt.plot(df_x['ra'], df_x['dec'],'pg', label= 'X')
plt.plot(df_gal['ra'], df_gal['dec'],'oy', label= 'Galaxy')
plt.plot(df_hb['ra'], df_hb['dec'],'+c', label= 'HB')
plt.plot(df_radio['ra'], df_radio['dec'],'Dg', label= 'Radio')
plt.plot(df_pm['ra'], df_pm['dec'],'Xc', label= 'PM')
plt.plot(df_wd_cand['ra'], df_wd_cand['dec'],'vr', label= 'Candidate_WD*')


plt.legend(loc = 'upper left', fontsize = 8)
plt.xlabel('RA (deg.)', fontsize = 12)
plt.ylabel('DEC (deg.)', fontsize = 12)
plt.tight_layout()
# plt.savefig('Figure/simbad_1', dpi=100)
```

```python
plt.figure(figsize=[7,5])
plt.plot(df_ttau['ra'], df_ttau['dec'],'*y', label= 'Ttau*')
plt.plot(df_HH['ra'], df_HH['dec'],'vm', label= 'HH')
plt.plot(df_cloud['ra'], df_cloud['dec'],'Xb', label= 'Part of Cloud')
plt.plot(df_radio['ra'], df_radio['dec'],'pc', label= 'Radio')
plt.plot(df_orian['ra'], df_orian['dec'],'Dm', label= 'Orian_V*')
plt.plot(df_IR['ra'], df_IR['dec'],'*k', label= 'IR')
plt.plot(df_radiosub['ra'], df_radiosub['dec'],'Pk', label= 'Radio(sub-mm)')
plt.plot(df_lmass['ra'], df_lmass['dec'],'xg', label= 'low-mass*')
plt.plot(df_dense['ra'], df_dense['dec'],'or', label= 'denseCore')
plt.legend(loc ='upper left', fontsize = 10)
plt.legend(loc='center left', bbox_to_anchor=(1, 0.5))
plt.xlabel('RA (deg.)', fontweight = 'bold', fontsize = 14)
plt.ylabel('DEC (deg.)', fontweight = 'bold', fontsize = 14)
plt.xticks(fontweight = 'bold', fontsize = 12)
plt.yticks(fontweight = 'bold', fontsize = 12)
plt.tight_layout()
# plt.savefig('Figure/simbad_2', dpi=100)
```
